```python
import glob
flist = glob.glob("airbnb/*.csv")
flist
```

```python
import pandas as pd
pd.read_csv(flist[0], encoding="utf-8")
```

```python
df = None
for fn in flist:
    temp = pd.read_csv(fn, encoding="utf-8")
    if df is None:
        df = temp
    else:
        df = pd.concat([df, temp], axis = 0, ignore_index=True)
df.to_csv("airbnb/reviews.csv", encoding="utf-8")
df
```

```python
sample = """Amazing flat, really close from the MAS Musem, and 15 minutes walking from the Cathedral. 

Really calm, many space and light, perfect for 2 couples, perfect for a weekend.
"""
from rake_nltk import Rake
r = Rake()
r.extract_keywords_from_text(sample)
r.get_ranked_phrases()
```

```python
def analyze(s):
    r = Rake()
    r.extract_keywords_from_text(s)
    return str(r.get_ranked_phrases())
# df["comments"].head(10).astype(str).apply(analyze)
df["key_phrase"] = df["comments"].astype(str).apply(analyze)
df.to_csv("key_phrase.csv", encoding="utf-8", index=False)
df
```
